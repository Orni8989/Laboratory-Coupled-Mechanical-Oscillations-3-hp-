```python
import numpy as np
import matplotlib.pyplot as plt

#Dessa är frekvenser för sqrt(g/l) tagna från simulering i "Labbplottar", det är en av de första simuleringarna och sedan de 4 längst ned
uppmatt_f1 = [0.799, 0.677, 0.838, 0.677, 0.612, 0.967]
teoretisk_f1 = [0.809, 0.681, 0.849, 0.666, 0.600, 0.959]

error = [0.032, 0.019, 0.037, 0.018, 0.013, 0.053] #insatta värden med felfortplantningsformeln, error längd till mc 3 cm, error längd g 0.01 m/s^2

bar_width = 0.35
index = np.arange(0, len(uppmatt_f1), 1)  # X-axelns positioner

# Skapa stapeldiagrammet
fig, ax = plt.subplots(figsize=(8, 6))

# Staplar för uppmätt frekvens
bars1 = ax.bar(index, uppmatt_f1, bar_width, yerr=error, capsize=5, label='Uppmätt frekvens med osäkerhet')

# Staplar för teoretisk frekvens
bars2 = ax.bar(index + bar_width, teoretisk_f1, bar_width, label='Teoretisk frekvens')

# Etiketter och titel
ax.set_xlabel('Mätserier', fontsize=16)
ax.set_ylabel('Frekvens (Hz)', fontsize=16)
#ax.set_title(r'Uppmätt och teoretisk frekvens för $f = \dfrac{\sqrt{\frac{g}{l}}}{2\pi}$', fontsize=20)
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_ylim(0.5, 1.1)
ax.set_yticks(np.arange(0.5, 1.1, 0.05))

# Lägg till en legend
ax.legend(loc = 'upper left', fontsize = 16)

# Visa plotten
plt.tight_layout()
plt.show()
```
